```python
from Bio.SeqIO.QualityIO import FastqGeneralIterator
import csv
import itertools
import sys
```

```python
sample = "Sample_N705_-_ER1"
project_path = "/home/arion/davidm/Data/datasets/raw/private/RNA-seq/sc/dsp779"

# project_path = sys.argv[1]
# sample = sys.argv[2]
```

```python
cbs = set(line.strip() for line in open(f"{project_path}//raw-fastqs/{sample}/CBs.txt"))
```

```python
cbs
```

```python
file = f"{project_path}/raw-fastqs/{sample}/{sample}_R1.fastq"
i=0
j=0
phreds = {cb : [] for cb in cbs}
umis = {cb : [] for cb in cbs}


with open(f"{project_path}/raw-fastqs/{sample}/new_phreds_test.csv", "w") as outfile:
    writer = csv.writer(outfile)
    writer.writerow(phreds.keys())
    
    
for _, seq, qual in FastqGeneralIterator(open(file)) :
    cb = seq[:12]
    if cb in cbs :
        phreds[cb].append(qual)
        umis[cb].append(seq[12:])
    i +=1
    if i > 1000000 :
        with open(f"{project_path}/raw-fastqs/{sample}/new_phreds_test.csv", "a") as outfile:
            writer = csv.writer(outfile)
            writer.writerow(phreds.values())
        phreds = {cb : [] for cb in cbs}
        umis = {cb : [] for cb in cbs}
        i = 0
        
    j +=1
    if j > 10000000 :
        break

with open(f"{project_path}/raw-fastqs/{sample}/new_phreds_test.csv", "a") as outfile:
    writer = csv.writer(outfile)
    writer.writerow(phreds.values())
```

```python
with open(f"{project_path}/raw-fastqs/{sample}/new_phreads_test.csv", "w") as outfile:
    writer = csv.writer(outfile)
    writer.writerow(phreds.keys())
    writer.writerows(itertools.zip_longest(*phreds.values()))
```

```python
with open(f"{project_path}/raw-fastqs/{sample}/new_phreads_test.csv", "w") as outfile:
    writer = csv.writer(outfile)
    writer.writerow(phreds.keys())
    writer.writerow(phreds.values())
```
